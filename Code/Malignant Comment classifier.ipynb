```python
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import  stopwords
import string
```

```python
training['length'] = training['comment_text'].str.len()
training.head(10)
```

```python
# Convert all messages to lower case
training['comment_text'] = training['comment_text'].str.lower()

# Replace email addresses with 'email'
training['comment_text'] = training['comment_text'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$',
                                 'emailaddress')

# Replace URLs with 'webaddress'
training['comment_text'] = training['comment_text'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',
                                  'webaddress')

# Replace money symbols with 'moneysymb' ( can by typed with ALT key + 156)
training['comment_text'] = training['comment_text'].str.replace(r'|\$', 'dollers')
    
# Replace 10 digit phone numbers (formats include paranthesis, spaces, no spaces, dashes) with 'phonenumber'
training['comment_text'] = training['comment_text'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$',
                                  'phonenumber')

    
# Replace numbers with 'numbr'
training['comment_text'] = training['comment_text'].str.replace(r'\d+(\.\d+)?', 'numbr')


training['comment_text'] = training['comment_text'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in string.punctuation))

stop_words = set(stopwords.words('english') + ['u', '', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure'])
training['comment_text'] = training['comment_text'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in stop_words))

lem=WordNetLemmatizer()
training['comment_text'] = training['comment_text'].apply(lambda x: ' '.join(
 lem.lemmatize(t) for t in x.split()))
```
