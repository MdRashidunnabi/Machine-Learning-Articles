# Feature Engineering 

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
```

```python
# returns the top 5 features in terms of correlation 
def rfe(training_feature):
    no_check = ['year', 'weekofyear', 'week_start_date']
    training = training_feature.drop(no_check, axis=1)
    x = training.loc[:,training.columns!='total_cases']
    y = training['total_cases']
    estimator_ = SVR(kernel="linear")
    selector_ = RFE(estimator_, 5, step=1)
    selector_ = selector_.fit(x, y)

    rank = selector_.ranking_
    col = x.columns.values
    for i in range(len(rank)):
        print(rank[i], col[i])
    return col[selector_.support_]
```

```python
# selecting for sj
print('sj feature ranks')
sj_rfe = rfe(sj_dengu)
```

```python
# selecting for iq
print('Ranking of iq')
iq_rfe = rfe(iq_dengu)
```

```python
clim_col = ['reanalysis_dew_point_temp_k',
                   'reanalysis_precip_amt_kg_per_m2',
                   'reanalysis_specific_humidity_g_per_kg',
                   'station_avg_temp_c', 
                   'station_max_temp_c',
                   'station_min_temp_c']
clim_col
```

```python
# adds rolling features
def adding_roll_feature(dff, windows):
    mean_col = ['reanalysis_dew_point_temp_k',
                    'reanalysis_specific_humidity_g_per_kg',
                    'station_avg_temp_c', 
                    'station_max_temp_c',
                    'station_min_temp_c']
    sum_col = ['reanalysis_precip_amt_kg_per_m2']
    
    roll_mean = dff[mean_col].rolling(window=windows, min_periods=1).mean()
    roll_sum = dff[sum_col].rolling(window=windows, min_periods=1).sum()

    roll_mean_col = ['roll_mean_' + column for column in mean_col]
    roll_sum_col = ['roll_sum_' + column for column in sum_col]

    roll_mean.columns = roll_mean_col
    roll_sum.columns = roll_sum_col

    dff = pd.concat([dff, roll_mean, roll_sum], axis=1)
    return dff
```

```python
# keeping track of the rolling columns
rolling_col = ['roll_mean_reanalysis_dew_point_temp_k',
                   'roll_sum_reanalysis_precip_amt_kg_per_m2',
                   'roll_mean_reanalysis_specific_humidity_g_per_kg',
                   'roll_mean_station_avg_temp_c', 
                   'roll_mean_station_max_temp_c',
                   'roll_mean_station_min_temp_c']
print('rolling_column')
rolling_col
```

```python
# adding rolling features to create new data set
sj_dengue_virus = adding_roll_feature(sj_dengu, 52)
iq_dengue_virus = adding_roll_feature(iq_dengu, 52)

# performing the same on the test set
sj_testing = adding_roll_feature(sj_testing, 52)
iq_testing = adding_roll_feature(iq_testing, 52)
```

# Using GridSearchCV to find the best parameter

```python
# returns model
def getting_md(x, y, estimator, param_grid):            
    
    grd_src = GridSearchCV(estimator=estimator, 
                               param_grid=param_grid, 
                               scoring="neg_mean_absolute_error", 
                               n_jobs=1, 
                               cv=KFold(n_splits=10, shuffle=True, random_state=1))
    
    grd_rslt = grd_src.fit(X=x, y=y)
    print(grd_rslt.best_score_)
    print(grd_rslt.best_params_)
    return grd_rslt
```

```python
# saves the submission into csv
def save(sj_predicts, iq_preditcs, file):
    submit = pd.read_csv("Submit.csv",
                             index_col=[0, 1, 2])
    submit.total_cases = np.concatenate([sj_predicts, iq_preditcs])
    save("data/" + file + ".csv")
```

```python
# plots time series of predicted vs actual cases
def plots_timing(train_cases, test_cases, actual_cases):
    total = pd.DataFrame()
    total['pred_cases'] = train_cases + test_cases
#     total['actual_cases'] = actual_cases
    
    # plotting time forecast of predicted cases vs actual cases
    plt.figure(figsize=(20, 10))
    total.pred_cases.plot(label="Prediction", color='C0')
    actual_cases.plot(label="Actual Data", color='C3')

    plt.ylabel('Counting Dengu')
    plt.title("Predicted_ Dengu Case vs. Actual_ Case", fontsize=20)
    plt.legend(fontsize=20)
```

```python
# plots prediction vs actual total cases values
def plotting_scatter(pred_cases, actual_cases):
    total = pd.DataFrame()
    total['pred_cases'] = pred_cases
    total['actual_cases'] = actual_cases
    plt.figure(figsize=(15,10))
    sns.regplot(x='pred_cases', y='actual_cases', data=total)
    plt.title('Predicting and Original Total Cases')
```
